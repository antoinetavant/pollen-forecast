```python
import pandas as pd
filename_deps = "../data/departements.geojson"
filename_cities = "../data/georef-france-commune.csv"
```

```python
df_cities = pd.read_csv(filename_cities)
df_cities.head()
```

```python
import json
with open(filename_deps, "r", encoding="UTF-8") as f:
    deps = json.load(f)
```

```python
deps_names = [feature["properties"]["nom"] for feature in deps["features"]]
deps_names
```

```python
for dep_name in deps_names:
    if dep_name not in df_cities["Nom Officiel Département"].values:
        print(dep_name)
```

```python
for dep_name in df_cities["Nom Officiel Département"].values:
    if dep_name not in deps_names:
        print(dep_name)
```
